# Data Preparation


```
 import pandas as pd
 
```

```
# we imported a data: Selling price of a given house in the Boston(MA) Area
from sklearn.datasets import load_boston
# we rename our dataset to boston_dataset
boston_dataset = load_boston()
boston_dataset
#boston_dataset.features
# we use the format to choose the dataset we want to use

```

```
#print the data row input  values.
print('data is :{}'.format(boston_dataset.data))
```

```
#print the independent variable in names ,the string names of the variables.
print('indpendent variable names are:{}'.format(boston_dataset.feature_names))

```

```
# print the dependent  value(The response variable)
print('labels or price:{}'.format(boston_dataset.target))
```

```
# puting your data in dataframe (in a table form): Panda helps u virsualize data more friendly way
df = pd.DataFrame(boston_dataset.data,columns = boston_dataset.feature_names)
# we create a new  column and  named  price and assign to target
df['Price'] = boston_dataset.target
# showing be the fist 10 datasets
df.head(n=10)
```

```
# Use the describe () method to understand the data.Helps to find if there is missing data.
df.describe(include ='all') 
```

```
#Create train and test databases
# this library helps you split your Data set.
from sklearn.model_selection import train_test_split
# we extract all the columns who does not have the name price
X = df.loc[:, df.columns != 'Price']
# we extract all the  column with the  name price
y = df.loc[:, df.columns == 'Price']
# random_state is similar to setting a setting a  seed in R
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```



When you standardize the whole dataset , it means that you are influencing the validation set with information from your learning data.
Hence, usually we use the mean and the standard deviation on the training data 
set to do standardization.




```
#Standardization
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)
X_train = (X_train - mean) / std
X_test = (X_test - mean) / std
```

# Model  Building

Architure based on two Densed layers. the first with 128 and the second with 64 neurons ,both using a ReLU(Rectified Linear UNit) activation function. A dense layer with a linear activation will be used as output layer.

Note: Since it a regression problem:  we need one output layer so only one node that is price
Steps: 
1. First import all the libararies you need 
2. You create an architure for your model


```
#import libarires 
from keras.models import Sequential
#'Dense' refers to a fully-connected layer
from keras.layers import Dense
#  creating the architure
#The Sequential model is the simplest model type --> a linear stack of layers
model = Sequential()
# To stack the layers, we use the add() method
# input_shape is mandotory for the first layer. the independent variable
#It means 13 input parameters, with 128 neurons in the FIRST hidden layer.
model.add(Dense(128, input_shape=(13,), activation ='relu', name ='dense_layer_1'))
model.add(Dense(64, activation ='relu', name ='dense_layer_2'))
model.add(Dense(1,  activation = 'linear', name ='dense_output'))
# Adding optimizer  and loss function 
#The compile() method allows us to configure the learning further
model.compile(optimizer='adam',loss='mse')
# if you are able to get a summary then you dont have bugs
model.summary()

```

```
# Now we Train our model:  fit our model using the train data
#
model.fit(X_train, y_train,epochs=100)

```

```
#Model evaluation 
loss = model.evaluate(X_test,y_test)

print("Mean squared Error on test data:",loss)


#
# Here we see that the Loss (mse) in the test data set is higher than the train data set hence our model overfit
```
